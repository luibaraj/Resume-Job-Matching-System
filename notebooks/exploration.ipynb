```python
import sqlite3
from pathlib import Path

# Connect to the database
db_path = Path("../data/jobs.db")
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# Get the total number of jobs
cursor.execute("SELECT COUNT(*) FROM jobs")
total_jobs = cursor.fetchone()[0]
print(f"Total number of jobs: {total_jobs}")

# Get the number of preprocessed jobs
cursor.execute("SELECT COUNT(*) FROM jobs WHERE preprocessed = 1")
preprocessed_jobs = cursor.fetchone()[0]
print(f"Preprocessed jobs: {preprocessed_jobs}")

# Get the number of extracted jobs
cursor.execute("SELECT COUNT(*) FROM jobs WHERE extracted = 1")
extracted_jobs = cursor.fetchone()[0]
print(f"Extracted jobs: {extracted_jobs}")

# Get the number of embedded jobs
cursor.execute("SELECT COUNT(*) FROM jobs WHERE embedded = 1")
embedded_jobs = cursor.fetchone()[0]
print(f"Embedded jobs: {embedded_jobs}")

# Get the number of embedded jobs
cursor.execute("SELECT COUNT(*) FROM jobs WHERE embedded = 1")
embedded_jobs = cursor.fetchone()[0]
print(f"Embedded jobs: {embedded_jobs}")

# Get the number of jobs with null or empty raw descriptions
cursor.execute("""
SELECT 
  COUNT(*) as total_jobs,
  COUNT(CASE WHEN raw_description IS NULL THEN 1 END) as null_descriptions,
  COUNT(CASE WHEN raw_description = '' THEN 1 END) as empty_descriptions,
  COUNT(CASE WHEN raw_description IS NOT NULL AND raw_description != '' THEN 1 END) as with_descriptions
FROM jobs
""")
result = cursor.fetchone()
print(f"Total jobs: {result[0]}")
print(f"Null descriptions: {result[1]}")
print(f"Empty descriptions: {result[2]}")
print(f"Jobs with descriptions: {result[3]}")

# Get jobs that failed preprocessing (have raw_description but no cleaned_description)
cursor.execute("""
SELECT COUNT(*) FROM jobs 
WHERE raw_description IS NOT NULL AND raw_description != '' 
AND cleaned_description IS NULL AND preprocessed = 0
""")
failed_cleaning = cursor.fetchone()[0]
print(f"Jobs with raw desc but no cleaned desc: {failed_cleaning}")

# Compare to preprocessed
cursor.execute("SELECT COUNT(*) FROM jobs WHERE preprocessed = 1")
preprocessed = cursor.fetchone()[0]
print(f"Successfully preprocessed: {preprocessed}")

conn.close()
```
